```python
#!/usr/bin/env python3

##########################################################
## For a given list of proteins the script resolves them
## (if possible) to the best matching STRING identifier
## and prints out the mapping on screen in the TSV format
##
## Requires requests module:
## type "python -m pip install requests" in command line
## (win) or terminal (mac/linux) to install the module
###########################################################

import requests ## python -m pip install requests
from time import sleep
```

# Format Genes for each Bacteria

```python
## genes should be a list

bacteria_genes = {}
gene_list = []
with open("../output/without_men/msea/msea_results.csv") as file:
    next(file)
    for line in file:
        print(line)
        gene = line.strip().split(",")[0]

        bacteria = line.strip().split("[\'")[1].split("\']")[0].replace("\'","").split(",")
        #bacteria = bacteria[1].split("\"]")
        #print("\"[")
        print("------------------------")
        #print(line)
        #print(line[0]) gene
        print(len(bacteria))
        #print(gene)
        if len(bacteria) == 3:
            gene_list.append(gene)
        for bacterium in bacteria: 
            bacterium = bacterium.strip()
            if bacterium not in bacteria_genes:
                bacteria_genes[bacterium] = [gene]
                #print(bacterium)
                #print(bacteria_genes[bacterium])
                #print("----------------------")
            else:
                #print("------------bacteria in the dict----------")

                gene_list = bacteria_genes[bacterium] + [gene]
                bacteria_genes[bacterium] = gene_list
                #print(bacterium)
                #print(bacteria_genes[bacterium])
                

print(bacteria_genes)
        
    

```

# Get STRING text output
get the string api url


```python
string_api_url = "https://version-12-0.string-db.org/api"
output_format = "tsv"
method = "network"# "get_string_ids"
my_genes = ["p53", "BRCA1", "cdk2", "Q99835"]
```

Get the parameters

```python
params = {

    "identifiers" : "\r".join(my_genes), # your protein list
    "species" : 9606, # NCBI/STRING taxon identifier 
    "limit" : 1, # only one (best) identifier per input protein
    "echo_query" : 1, # see your input identifiers in the output
    "caller_identity" : "www.awesome_app.org" # your app name

}
```

Construct URL

```python
request_url = "/".join([string_api_url, output_format, method])
request_url
```

Call STRING

```python
results = requests.post(request_url, data=params)

results.text
```

Read and parse the results

```python
for line in results.text.strip().split("\n"):
    print(f"{line}\n")
    line = line.split("\t")
    #print(line)
    #l = line.split("\t")
    #print(l)
    #input_identifier, string_identifier = l[0], l[2]
    #print("Input:", input_identifier, "STRING:", string_identifier, sep="\t")
    
```

# Getting STRING network image 


```python
string_api_url = "https://version-12-0.string-db.org/api"
output_format = "image"
method = "network"

##
## Construct URL
##


request_url = "/".join([string_api_url, output_format, method])

## For each gene call STRING

for gene in my_genes:

    ##
    ## Set parameters
    ##

    params = {

        "identifiers" : gene, # your protein
        "species" : 4932, # NCBI/STRING taxon identifier 
        "add_white_nodes": 15, # add 15 white nodes to my protein 
        "network_flavor": "confidence", # show confidence links
        "caller_identity" : "www.awesome_app.org" # your app name

    }


    ##
    ## Call STRING
    ##

    response = requests.post(request_url, data=params)

    ##
    ## Save the network to file
    ##

    file_name = f"../output/without_men/{gene}_network.png"
    print("Saving interaction network to %s" % file_name)

    with open(file_name, 'wb') as fh:
        fh.write(response.content)

    sleep(1)
```

```python
string_api_url = "https://version-12-0.string-db.org/api"
output_format = "image"
method = "network"
request_url = "/".join([string_api_url, output_format, method])

params = {

    "identifiers" : "%0d".join(my_genes), # your protein
    "species" : "Homo sapiens", # NCBI/STRING taxon identifier 
    "add_white_nodes": 0, # add 15 white nodes to my protein 
    "network_flavor": "evidence", # show confidence links
    "caller_identity" : "www.awesome_app.org", # your app name
    "show_query_node_labels": 1

}


##
## Call STRING
##

response = requests.post(request_url, data=params)

##
## Save the network to file
##

file_name = f"multiple_genes_network.png"
print("Saving interaction network to %s" % file_name)

with open(file_name, 'wb') as fh:
    fh.write(response.content)

sleep(1)
```
